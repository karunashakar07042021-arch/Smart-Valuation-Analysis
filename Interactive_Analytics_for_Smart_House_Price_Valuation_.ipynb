**Check skewness and apply transformations**

```python
df["Sale_Price"].skew()
```

**What it means:**
- Skewness measures how asymmetric a distribution is compared to a normal (bell‑shaped) distribution.

- A skewness of 0 means the data is perfectly symmetric.

- Positive skew (>0) means the distribution has a long tail to the right (values stretch higher).

- Negative skew (<0) means the distribution has a long tail to the left (values stretch lower).


| Column | Skewness | Interpretation |
| --- | --- | --- |
| Sale_Price | 2.660964 | Strong right‑skew (typical in housing prices, with a few very expensive homes). |

```python
df_log = np.log1p(df["Sale_Price"])
df_log.skew()
```

**Skewness measures how asymmetric the distribution is:**

- 0 → perfectly symmetric (like a normal distribution).

- Positive skew → long right tail (common in housing prices, lot sizes).

- Negative skew → long left tail.

**Applied log transformation using np.log1p() to reduce skewness**

| Column_Name | Skewness | Interpretation |
